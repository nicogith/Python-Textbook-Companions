## Example 11 , Page number: CP-94

## Example 15 , Page number: CP-99

## Example 17 , Page number: CP-102

```python
import math

# Program to compute the value of pie
# Variable declaration

s = 0.0
dr = 1
sign = 1
term = (1.0/dr) * sign
while (math.fabs(term) >= 1.0e-4):
    s = s + term
    dr = dr + 2
    sign = sign * -1
    term = (1.0 / dr) * sign

pie = s * 4

print "Value of pie is %.6f"%pie
```
