## EXAMPLE ON PAGE:434-435

```python
 
import sys
if(len(sys.argv)!=3):
    print "Improper number of arguments\n"
    sys.exit(0)
try:
    fs=open(sys.argv[1], "r")
except:
    sys.exit("Cannot open source file\n")
try:
    ft=open(sys.argv[2], "w")
except:
    fs.close()
    sys.exit("Cannot open source file\n")
ch=fs.readlines()
ft.writelines(ch)
fs.close()
ft.close()
```
