## Example 4.4, page no. 109

```python

def sq_root(x):
    DELTA = 0.0001
    curr_appx = 0
    last_appx = 0
    diff = 0
    last_appx = x
    diff = DELTA+1
    while(diff > DELTA):
        curr_appx = 0.5*(last_appx + x/last_appx)
        diff = curr_appx - last_appx
        if(diff < 0):
            diff = -diff
            last_appx = curr_appx
    return(curr_appx)
 


for i in range(1,101):
    print "root of %d is %f\n" %( i, sq_root(i))
```
