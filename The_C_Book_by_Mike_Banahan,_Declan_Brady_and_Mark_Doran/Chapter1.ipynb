## Example 1.2, page no. 22

```python

this_number = 3
divisor = 0 
not_prime = 0

while(this_number < 10000):
    divisor = this_number / 2
    not_prime = 0
    while(divisor > 1):
        if(this_number % divisor == 0):
            not_prime = 1
            divisor = 0
        else:
            divisor = divisor-1

    if(not_prime == 0):
        print "%d is a prime number\n" %this_number 
    this_number = this_number + 1;
```
