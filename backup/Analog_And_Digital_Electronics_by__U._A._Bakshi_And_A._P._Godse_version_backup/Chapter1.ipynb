## Example 1.2 Page No. 1-20

```python
import math

# Formula
print("The transistor capacitance is given by,");                              
print("CT = C(0)/[1+|VR/VJ|^n]\n");

# Given Data
print("Now C(0) = 80pF , n = 1/3 as diffused junction");                      
print("VR = 4.2 V, VJ = 0.7 V\n");

# Calculation in pF
ct = round((80*pow(10,-12))/((pow((1+(4.2/0.7)),(0.3333))))*pow(10,12),2);     # Note here, 1/3 = 0.3333 

# Result
print "Therefore, CT(pF) = ",ct,"pF";         

# Formula
print("\nThe transistor capacitance is also given by, " )
print("CT = K/[VR+VJ]^n\n")       

#Calculation
k = round((41.82*pow(10,-12))*(pow((4.2+0.7),(0.3333)))*pow(10,12),2);         # Note here, 1/3 = 0.3333 

# Result
print "Therefore, K = ",k,"* 10^-12";                                                    
```
