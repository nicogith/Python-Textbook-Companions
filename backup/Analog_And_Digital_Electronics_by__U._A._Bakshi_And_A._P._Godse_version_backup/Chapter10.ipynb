## Example 10.1 Page No. 10-13

```python
import math

print("ZZ = 7ohm,R3 = 330ohm,V0 = 4.7V,Vin = 15V")
print("The specified change in Vin is 10%,")
vin = 0.1*15
print "Therefore,delta Vin(in V) = 10% of Vin=",vin,"V"
vo = (1.5*7)/330
print "Therefore,delta V0(in V) = delta Vin∗ZZ/R3=",round(vo,5),"V"
lr = 0.03181*100/4.7
print "Therefore,Line regulation(in percentage) = deltaV0*100/V0=",round(lr,3),"%"
print("For IL(max) = 50mA,")
dvo=(20.0*7.0*50.0*pow(10,-3))/330.0
print "Therefore,delta V0(in V) = IL(max)∗RS∗ZZ/R3=",round(dvo,5),"V"
lr=0.02121*100/4.7
print "Therefore,Line regulation(in precentage) = deltaV0∗100/V0=",round(lr,4),"%"
print("Now VR(out) = VR(in)∗ZZ/R3")
zz=7.0/330.00
print "Therefore,VR(out)/VR(in) = ZZ/R3 = ",round(zz,5)
rr=20*math.log10(0.02121)
print "Therefore,RR (in dB) = 20∗log(0.02121)=",round(rr,2),"dB"
```

## Example 10.2 Page No. 10-15 

```python
import math

print("Vo = (1+R1/R2)Vz")
print("9 = (1+R1/R2)5.6")
print("Therefore R1/R2 = 0.6071")
print("Let R1 = 10KOhm")
print("Therefore R2 = 16.47KOhm")
print("Minimum zener current is say 5 mA, which is same as current through R3")
print("Therefore Izmin = (Vinmin - Vz)/R3")
print("Therefore 5x10^-3 = ([18-3] - 5.6)/R3")
print("R3 = 1.88KOhm")
print("For Vmax = 18+3 = 21V we get,")
print("Iz = (21-5.6)/1.88x10^3 = 8.19 mA")
print("Pd = IzxVz = 8.19x10^-3 x 5.6 = 0.045W")
print("Thus Pd < Pzmax")
print("Thus the designed elements are")
print("R1 = 10KOhm")
print("R2 = 16.47KOhm")
print("R3 = 1.88KOhm")
print("Assume Beta of the series transistor as 30")
print("IB = IE/(Beta +1) = Iomax/(Beta + 1)")
print("IB = 50/31 = 1.61 mA")
print("The Op-amp must be capable of supplying this current. The op-amp 741 has output current rating of 25mA. Hence op-amp 741 can be used")
print("")
```

```python
import math

print("The (Io)max of 7805 regulator is 1A")
print("IL = (1+B)Io - B(Vbe/R1)")
print("(IL)max = (1+B)(Io)max - B(Vbe/R1)")
print("(IL)max = (1+15)x1 - (15 x 1/7) = 13.85 A")
print("Thus IC which can supply maximum of 1A can supply maximum load of 13.85A, with the help of the current boosting arrangement")
```
